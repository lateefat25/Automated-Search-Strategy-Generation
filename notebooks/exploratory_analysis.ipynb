```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

```

```python
# Load Dataset
df = pd.read_csv("/home/lateefat/Automated Search Strategy Generation/data/job_descriptions.csv")  # Update path if necessary
```

```python
# Display basic info
print("Dataset Overview:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Job Title Distribution
plt.figure(figsize=(12, 6))
sns.countplot(y=df['Job Title'], order=df['Job Title'].value_counts().index[:10])
plt.title("Top 10 Job Titles")
plt.xlabel("Count")
plt.ylabel("Job Title")
plt.show()
```

```python
# Skill Frequency Analysis
def plot_wordcloud(column, title):
    text = " ".join(df[column].dropna())
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(title)
    plt.show()

# Generate word clouds for skills and job descriptions
plot_wordcloud('skills', "Most Common Skills")
```

```python
plot_wordcloud('Job Description', "Job Description Word Cloud")
```

```python
# Analyze Description Length
df['Description_Length'] = df['Job Description'].astype(str).apply(len)
plt.figure(figsize=(10, 5))
sns.histplot(df['Description_Length'], bins=30, kde=True)
plt.title("Distribution of Job Description Lengths")
plt.xlabel("Character Count")
plt.ylabel("Frequency")
plt.show()
```
